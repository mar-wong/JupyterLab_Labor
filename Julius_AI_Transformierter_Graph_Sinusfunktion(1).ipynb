```python
# Plot the transformed sine function per the description
import numpy as np
import matplotlib.pyplot as plt

# Define base and transformed functions
x_vals = np.linspace(-2*np.pi, 2*np.pi, 1000)

def g(x):
    return np.sin(x)

# Transformations: stretch by 3 in y, shift right by pi/2, shift down by 2
# f(x) = 3 * sin(x - pi/2) - 2

def f(x):
    return 3*np.sin(x - np.pi/2) - 2

# Prepare figure
plt.figure(figsize=(8,4.5))
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# Plot base sine for reference (light gray)
plt.plot(x_vals, g(x_vals), color='gray', alpha=0.4, label='g(x) = sin(x)')
# Plot transformed function
plt.plot(x_vals, f(x_vals), color='#1f77b4', label='f(x) = 3 sin(x - Ï€/2) - 2')

# Annotate key features
# amplitude, midline, phase shift
plt.axhline(-2, color='#1f77b4', linestyle='--', linewidth=1, alpha=0.7, label='Mittellinie y = -2')

# Mark one period key points for f: start at x = -pi/2 gives -2 (midline), max at x = 0, min at x = pi, etc.
key_x = np.array([ -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2 ])
key_y = f(key_x)
plt.scatter(key_x, key_y, color='#1f77b4')
for xv, yv in zip(key_x, key_y):
    plt.text(xv, yv+0.3, "(%.2f, %.2f)" % (xv, yv), ha='center', va='bottom', fontsize=8)

# Formatting
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-5, 3)
xticks = [-2*np.pi, -3*np.pi/2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xtick_labels = ["-2Ï€","-3Ï€/2","-Ï€","-Ï€/2","0","Ï€/2","Ï€","3Ï€/2","2Ï€"]
plt.xticks(xticks, xtick_labels)
plt.yticks(range(-5,4))
plt.legend(loc='upper right')
plt.title('Transformation von g(x)=sin(x) zu f(x)')
plt.tight_layout()
print("Plotted f(x) = 3 sin(x - Ï€/2) - 2 und g(x) = sin(x)")
```
