```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 0.3
g = 9.81
dt = 0.01
t = np.arange(0, 2+dt, dt)
theta = np.zeros(len(t))
thetadot = np.zeros(len(t))
theta[0] = 45*np.pi/180
for i in range(0, len(t)-1):
    thetaddot = -g*np.sin(theta[i])/L
    thetadot[i+1] = thetadot[i] + thetaddot*dt
    theta[i+1] = theta[i] + thetadot[i+1]*dt

plt.plot(t, theta)
plt.title('Numerische Lösung der DLG des Fadenpendels')
plt.text(0.1, 0.65, r'DLG: $L\cdot \ddot{\theta \ } +\ g\cdot sin( \theta ) =\ 0$')
plt.xlabel('time (s)')
plt.ylabel('theta (rad)')
plt.show()
```
