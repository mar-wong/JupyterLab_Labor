<font color='red'>**Numerische Methode für
Differentialgleichung eines idealen Federpendels** </font>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 0.3
k = 4
dt = 0.01
t = np.arange(0, 2+dt, dt)
elong = np.zeros(len(t))
elongdot = np.zeros(len(t))
elong[0] = 0.1
for i in range(0, len(t)-1):
    elongddot = -k*elong[i]/m
    elongdot[i+1] = elongdot[i] + elongddot*dt
    elong[i+1] = elong[i] + elongdot[i+1]*dt

plt.plot(t, elong)
plt.grid()
plt.title('Numerische Lösung der DLG des Federpendels')
plt.text(0.5, 0.05, r'DLG: $m\cdot \ddot{x\ } +\ D\cdot x=\ 0$')
plt.xlabel('time (s)')
plt.ylabel('Elongation (m)')
plt.show()
```

<font color='blue'>Numerische Lösung der Differentialgleichung eines realen Federpendels </font>

Dämpfung wird betrachtet: $F_{Reib} =-c\cdot \dot{x}$

<font color='blue'> Differentialgleichung ist:</font>

$m\cdot \ddot{x\ } +c\cdot \dot{x} \ +D\cdot x=\ 0$

Man kann verschiedene Werte für Masse, Federkonstante, Dämpfungsfaktor und Iteration des Zeitintervalls in Code eingeben und entsprechenden Änderungen im Graphen der Elongation als Funktion der Zeit beobachten.


```python
m = 0.3
k = 4
c = 0.8
dt = 0.01
t = np.arange(0, 10+dt, dt)
elong = np.zeros(len(t))
elongdot = np.zeros(len(t))
elong[0] = 0.1
for i in range(0, len(t)-1):
    elongddot = -k*elong[i]/m - c*elongdot[i]
    elongdot[i+1] = elongdot[i] + elongddot*dt
    elong[i+1] = elong[i] + elongdot[i+1]*dt

plt.plot(t, elong)
plt.grid()
plt.show()
```
