```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from ipywidgets import interact, FloatSlider, IntSlider, FloatText
```

```python
#numerische LÃ¶sung der ODE zum Fadenpendel mit Sliders
def rightSideODE(x,t,k1,k2):
    dxdt=[x[1],-k1*x[1]-k2*np.sin(x[0])]
    return dxdt
# set the initial conditions
x0=[0,1]
 
# define the discretization points
timePoints=np.linspace(0,100,300)

def plot_func(k1, k2):
    solutionOde=odeint(rightSideODE,x0,timePoints, args=(k1,k2))
    plt.plot(timePoints, solutionOde[:, 0], 'b', label='x1')
    #plt.plot(timePoints, solutionOde[:, 1], 'g', label='x2')
    plt.legend(loc='best')
    plt.xlabel('time')
    plt.ylabel('x1(t), x2(t)')
    plt.grid()
    #plt.savefig('simulation.png')
    plt.show()
    
k1_slider = FloatSlider(value=0.4, min=0.4, max=2, step=0.01, description='k1:')
k2_slider = FloatSlider(value=2, min=2, max=6, step=0.01, description='k2:')
interact(plot_func, k1=k1_slider, k2=k2_slider)
pass
```
